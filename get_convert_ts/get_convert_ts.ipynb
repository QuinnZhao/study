```python
import requests

def get_ts_list(url):
    # 获取m3u8文件并提取其中的ts文件列表
    m3u8_txt = requests.get(url, headers = {'Connection':'close'}).text
    ts_list = []
    _l = m3u8_txt.split('\n')
    for item in _l:
        if '.ts?' in item:
            ts_list.append(item)
    return ts_list 

def get_ts_files(ts_list, url_root):
    # 下载ts文件
    if not ts_list:
        return
    import time
    for index,item in enumerate(ts_list):
        url = url_root + item
        # file_name = 'media_' + str(index + 1)+ '.ts'
        file_name = 'media_{:03d}.ts'.format(index+1)
        with open(file_name, 'wb') as f_out:
            print('writing {}'.format(file_name))
            movie = requests.get(url, headers = {'Connection':'close'}, timeout=60)
            f_out.write(movie.content)
        time.sleep(1)
```
